Now I will try to vizualize the what is the ratio between positive and negative diagnosis.

First I will create dataframe with only the patient ID and diagnosis.


```python
diagnosis_data = asthma_data[["PatientID","Diagnosis"]]
```

```python
diagnosis_data["Diagnosis"]
```

```python
positive = diagnosis_data.drop(diagnosis_data[diagnosis_data['Diagnosis'] == 1].index).value_counts().sum()
```

```python
positive
```

```python
negative = diagnosis_data.drop(diagnosis_data[diagnosis_data['Diagnosis'] == 0].index).value_counts().sum()
```

```python
negative
```

```python
data = {
    'Type': ['Positive', 'negative'],
    'Value': [positive , negative]
}
plot_data = pd.DataFrame(data)

# Create the bar plot using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Type', y='Value', data=plot_data)
plt.title('Positive vs Negative Diagnosis')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
negative/positive*100
```

Only 5.5 % of the patients are with negative diagnosis.
